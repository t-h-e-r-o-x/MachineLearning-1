```python
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
from collections import Counter
import pandas as pd
import random
```

```python
df = pd.read_csv('breast-cancer-wisconsin.data.txt')

df.replace('?','-999999', inplace = True)
df.drop(['id'], 1, inplace = True)
#print(df)
full_data = df.astype(float).values.tolist()
#print(full_data)
```

```python
random.shuffle(full_data)
test_size = 0.2
train_set = {2:[], 4:[]}
test_set = {2: [], 4:[]}

train_data = full_data[0:-int(test_size*len(full_data))]
test_data = full_data[-int(test_size*len(full_data)):]

for data in train_data:
    train_set[data[-1]].append(data[0:-1])
for data in test_data:
    test_set[data[-1]].append(data[0:-1])
test_set
```

```python
def knn(data, predict, k=3):
    distances = []
    for group in data:
        for features in data[group]:
            d = np.linalg.norm(np.array(features) - np.array(predict))
            distances.append([d, group])
    votes = [i[1] for i in sorted(distances)[:k]]
    
    vote_result = Counter(votes).most_common(1)[0][0]
    return vote_result

new_pt = [4,1,4,1,2,1,1,1,1]
knn(train_set, new_pt, 500)
```

```python
c = 0
t = 0

for k in range(0,20): #Analysing how accuracy changes with the value of k
    for grp in test_set:
        for dat in test_set[grp]:
            #print(grp, end = '->')
            #print(dat, end = '->')
            res = knn(train_set, dat, 6)
            #print(res)
            if(grp == res):
                c += 1
                #print("In")
                #print(c)
            t += 1
            #print(t)
    print(k,'---->',end = ' ')
    print('Accuracy',(c/t))
```
